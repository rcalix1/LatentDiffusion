
## Noise, t, gaussian, marginal


```python

import torch
import numpy as np
```

```python


# batch of 3 images (each image = 1 pixel for simplicity)
# shape [B, C] instead of [B, C, H, W]
x = torch.tensor([[1.0],
                  [1.0],
                  [1.0]])   # shape [3,1]

x.shape
```

```python

## this is like the time sequence, but because we use batches it is not in order, 
## so 0.12 is earlier in time and less corrupted or noisy 
## whereas 0.88 is more corrupted and noisy
## as an anolog to time 1, 2, 3
## it would be 0.12, 0.23, 0.88
## but because of parallel and batching they can be out of sequence
## each t is trained independently 

# one time per image

t = torch.tensor([0.12, 0.88, 0.23])  # shape [3]
```

```python

## this is from diff equations 
## essentially it is a gaussian, and the t relates to the standard dev of the gaussian,
## the formula maps higher to more standard dev, and more noise 

sigma = 25.0

def marginal_prob_std(t, sigma):
    return torch.sqrt((sigma**(2*t) - 1.) / (2 * np.log(sigma)))
```

```python

std = marginal_prob_std(t, sigma)  # shape [3]

std.shape
```

```python

# noise per image
z = torch.tensor([[0.5],
                  [-1.0],
                  [0.2]])   # shape [3,1]

print( z.shape )
```

```python

# apply noise (broadcast std)

perturbed_x = x + z * std[:, None]
```

```python


print("x:", x.squeeze())
print("t:", t)
print("std:", std)
print("z:", z.squeeze())
print("perturbed_x:", perturbed_x.squeeze())
```
